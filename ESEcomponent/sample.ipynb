```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Load the dataset
df = pd.read_csv('new.csv')

# Change the column names
df.columns = df.columns.str.replace(' ', '_')

# Display the updated column names
print(df.columns)

# Convert the 'Review Text' column to strings
df['Review Text'] = df['Review Text'].astype(str)

# Define the stopwords
stop_words = set(stopwords.words('english'))
# Perform text preprocessing on the 'Review Text' column
df['Review Text'] = df['Review Text'].apply(lambda x: ' '.join([word.lower() for word in word_tokenize(x) if word.isalnum() and word not in stop_words]))


# Display the preprocessed 'Review Text' column
print(df['Review Text'])
# # Display the preprocessed 'Review Text' column
# print(df['Review_Text'])
```
